```python
# Modules

# To list files in a directory
import os

# Time for execution of the code
import time # for debugging

# For .json files
import json

# Count dictionary values
from collections import Counter

# Pick a random item from a list
import random

# Print dictionaries prettily
import pprint
```

```python
# List all the bills
#print(os.getcwd()) # for debugging
bills_path = "2023-2024_118th_Congress\\bill"
raw_bills_files_list = os.listdir(bills_path)
#print(raw_bills_files_list) # for debugging
print("There are "+str(len(raw_bills_files_list))+" bills.") # for debugging
```

```python
# Initialize the list of bill metadata and the timer
bills = []
total_time = 0
```

```python
# Gather data for each bill
for file in raw_bills_files_list:
    with open(bills_path+"\\"+file, 'r') as bill_file:
        # Start the timer
        start = time.time() # for debugging
        
        # Convert to a python object
        obj = json.load(bill_file)
        print("Loop is on "+obj["bill"]["bill_number"]+"...\n") # for debugging

        # Initialize a dictionary with the data I want about the bill
        bill = {
            "chamber":obj["bill"]["body"], 
            "bill_type":obj["bill"]["bill_type"],
            "title":obj["bill"]["bill_number"]+":  "+obj["bill"]["title"],
            "summary":obj["bill"]["description"],
            "introduced":obj["bill"]["history"][0]["date"]
        }

        # Passage in the chamber
        passage_vote = list(filter(lambda roll_call: roll_call["desc"].startswith("On Passage RC#"), obj["bill"]["votes"]))
        if len(passage_vote) == 0:
            bill["passage_vote"] = "No"
        else: 
            bill["passage_vote"] = "Yes"

        # Sponsor(s)
        # Get the people with their party and state
        bill["sponsor(s)"]={
            "people":[]
        }
        sps = obj["bill"]["sponsors"]
        for sponsor in sps:
            sponsor_index = sps.index(sponsor)
            with open('2023-2024_118th_Congress\\people\\'+str(sps[sponsor_index]["people_id"])+'.json', 'r') as voter_file:
                    voter_obj = json.load(voter_file)

                    # Append dictionary
                    bill["sponsor(s)"]["people"].append({
                        "name":voter_obj["person"]["name"],
                        "party":voter_obj["person"]["party"],
                        "state":voter_obj["person"]["district"][3:5]
                    })   
        # Add the number of each party
        c = Counter()
        for item in bill["sponsor(s)"]["people"]:
            c[item["party"]] += 1
        bill["sponsor(s)"].update(c)            

        # Add the subject(s) of the bill
        bill["subject(s)"]=[]
        for subject in obj["bill"]["subjects"]:
            bill["subject(s)"].append(subject["subject_name"])

        # Number of actions and latest action
        bill["actions"] = len(obj["bill"]["history"])
        bill["latest_action"] = obj["bill"]["history"][len(obj["bill"]["history"])-1]["action"]

        # Add a dictionary about the latest vote, if there is one
        vote = obj["bill"]["votes"]
        if vote == []:
            bill["latest_vote"]={
                "date":"",
                "vote_on":"",
                "votes":""
            }
        else:    
            latest_index = len(vote)-1
            bill["latest_vote"]={
                "date":vote[latest_index]["date"],
                "vote_on":vote[latest_index]["desc"]  
            }
            # Open the file with the roll_call_id
            with open('2023-2024_118th_Congress\\vote\\'+str(vote[latest_index]["roll_call_id"])+'.json', 'r') as vote_file:
                vote_obj = json.load(vote_file)

                # Append the votes dictionary
                bill["latest_vote"]["votes"] = {
                    "people":[]
                }
                # Fill in the people dictionaries
                votes = vote_obj["roll_call"]["votes"]
                for vote in votes:   
                    # Get and add the politician's name, party, and state
                    vote_index = votes.index(vote)
                    with open('2023-2024_118th_Congress\\people\\'+str(votes[vote_index]["people_id"])+'.json', 'r') as voter_file:
                        voter_obj = json.load(voter_file)

                        # See the contents of voter_file, formatted nicely
                        #voter_formatted_str = json.dumps(voter_obj, indent=4)
                        #print(voter_formatted_str)

                        bill["latest_vote"]["votes"]["people"].append({
                            # Append name
                            "name":voter_obj["person"]["name"],
                            "party":voter_obj["person"]["party"],
                            "state":voter_obj["person"]["district"][3:5],
                            "vote":votes[vote_index]["vote_text"]
                        })

                # Add the number of each party
                c_yea = Counter()
                c_nay = Counter()
                c_absent = Counter()
                for item in bill["latest_vote"]["votes"]["people"]:
                    if item["vote"] == "Yea":
                        c_yea[item["party"]] += 1
                    if item["vote"] == "Nay":
                        c_nay[item["party"]] += 1
                    if item["vote"] == "Absent":
                        c_absent[item["party"]] += 1    
                bill["latest_vote"]["votes"]["Yea"] = dict(c_yea)
                bill["latest_vote"]["votes"]["Nay"] = dict(c_nay)
                bill["latest_vote"]["votes"]["Absent"] = dict(c_absent)
        bills.append(bill)
        end = time.time() # for debugging
        total_time = total_time + end-start
        #print("Time elapsed for this bill:", end-start, "seconds\n") # for debugging
        print("Total time elapsed:", total_time, "seconds\n\n")
```

```python
# Select a random bill
selected_bill = random.choice(bills)
selected_index = bills.index(selected_bill)

# Print the original file
print("Selected the "+str(selected_index+1)+"th bill, "+selected_bill["title"].split(":")[0]+".\n")
print("Here is the original file:\n")
with open(bills_path+"\\"+bills[selected_index]["title"].split(":")[0]+".json", 'r') as bill_file:
    # Convert to a python object
    obj = json.load(bill_file)

    # See the contents of bill_file, formatted nicely
    json_formatted_str = json.dumps(obj, indent=4) 
    print(json_formatted_str) 

# Print my metadata
print("\nHere is my metadata:\n")
pprint.pprint(selected_bill, sort_dicts=False)
```
